```python
import numpy as np
f = False
def compute_mt_vt_grad_updates(grad_history, beta1=0.9, beta2=0.999, epsilon=1e-8, learning_rate=0.001):
    """
    Calculate the moving average (mt) and variance (vt) buffers using exponential moving average,
    apply bias correction, and compute the effective gradient update with the learning rate.

    Parameters:
    - grad_history: List of gradients over time.
    - beta1: Exponential decay rate for the first moment estimates (default: 0.9).
    - beta2: Exponential decay rate for the second moment estimates (default: 0.999).
    - epsilon: Small constant for numerical stability (default: 1e-8).
    - learning_rate: Learning rate for the optimizer (default: 0.001).

    Returns:
    - mt: Moving average of the gradients.
    - vt: Moving average of the squared gradients.
    - grad_updates: Effective gradient updates with learning rate.
    """
    mt = np.zeros_like(grad_history[0])
    vt = np.zeros_like(grad_history[0])
    grad_updates = []

    for t, grad in enumerate(grad_history, 1):

        if t == 5 and f == True:
            mt *=3
            vt *= 9
        if t>= 5 and f == True:
            grad*=3

        # Update biased first moment estimate
        mt = beta1 * mt + (1 - beta1) * grad

        # Update biased second raw moment estimate
        vt = beta2 * vt + (1 - beta2) * (grad ** 2)

        # Compute bias-corrected first moment estimate
        mt_hat = mt / (1 - beta1 ** t)

        # Compute bias-corrected second raw moment estimate
        vt_hat = vt / (1 - beta2 ** t)

        # Compute effective gradient update
        grad_update = (learning_rate * mt_hat) / (np.sqrt(vt_hat) + epsilon)

        grad_updates.append(grad_update)

    return mt, vt, grad_updates
```

```python
grads = np.random.normal(0,10,(10,))
print(grads)
```

```python
mt,vt ,grad_updates = compute_mt_vt_grad_updates(grads,learning_rate=1)
print(mt)
print(vt)
print(grad_updates)
```

```python
mt,vt ,grad_updates = compute_mt_vt_grad_updates(grads / 10,learning_rate=1)
print(mt)
print(vt)
print(grad_updates)
```
